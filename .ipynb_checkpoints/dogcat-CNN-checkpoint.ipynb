```python
import tensorflow as tf
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
TRAIN_DIR = "./dataset/train/"
size = 100
```

```python
import os
```

```python
train_x = []
train_y = []

for i in os.listdir(TRAIN_DIR):
    img = cv2.imread(os.path.join(TRAIN_DIR,i), 0)
    img = cv2.resize(img, dsize = (size,size))
    train_x.append(img)
    if i.split('.')[0] == 'cat':
        train_y.append(1)
    else:
        train_y.append(0)

train_x = np.array(train_x)
train_y = np.array(train_y)
train_y = tf.keras.utils.to_categorical(train_y)
```

```python
len(train_x)
```

```python
plt.imshow(train_x[0])
plt.show()
```

```python
train_x = train_x.reshape(25000,100,100,1)
train_x = train_x.astype('float32')/255

x_test = train_x[-5000:]
y_test = train_y[-5000:]

x_train = train_x[:-5000]
y_train = train_y[:-5000]
```

```python
model2 = tf.keras.models.Sequential()

model2.add(tf.keras.layers.Conv2D(64,(5,5), activation = tf.nn.relu, input_shape = (100,100,1)))
model2.add(tf.keras.layers.MaxPooling2D((2,2)))
model2.add(tf.keras.layers.Conv2D(128,(5,5), activation = tf.nn.relu))
model2.add(tf.keras.layers.MaxPooling2D((2,2)))
model2.add(tf.keras.layers.Conv2D(128,(5,5), activation = tf.nn.relu))
model2.add(tf.keras.layers.MaxPooling2D((2,2)))
model2.add(tf.keras.layers.Conv2D(64,(5,5), activation = tf.nn.relu))
model2.add(tf.keras.layers.MaxPooling2D((2,2)))
model2.add(tf.keras.layers.Flatten())
model2.add(tf.keras.layers.Dense(1024, activation = tf.nn.relu))
model2.add(tf.keras.layers.Dense(2, activation = tf.nn.softmax))
```

```python
model2.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

```python
model2.fit(x_train, y_train, epochs = 4, batch_size = 100)
```
